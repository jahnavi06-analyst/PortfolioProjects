```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Medium','High']
```

```python
df['Income_group']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df[['Estimated Income','Income_group']]
```

```python
df['Income_group'].value_counts().plot(kind='bar')
```

```python
df
```

```python
#Univariate analysis
```

```python
categorical_cols=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_group']].columns

for col in categorical_cols:
    print(f"value Counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i,predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_group']].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

```python
#Bivariate analysis
```

```python
for i,predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_group']].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
for i,predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_group']].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
numerical_col=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Foreign Currency Account']
correlation_matrix=df[numerical_col].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title('correlation_matrix')
plt.show()
```
